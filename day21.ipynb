```python
m = s.split()
sr, sc = 0,0
for i, l in enumerate(m):
    if 'S' in l:
        sr, sc = i, l.index('S')
        break

dmaps = {}
R = len(m)
assert len(m) == len(m[0])
for st in (sr,sc), (0,0), (0,R-1), (sr,0), (sr,R-1), (R-1,0), (R-1,R-1), (0,sc), (R-1, sc):
    dmap = {}
    q = [(0,st)]
    while q:
        d, cur = q.pop(0)
        if cur in dmap: continue
        dmap[cur] = d
        for r,c in neighbors4(*cur):
            if not (0 <= r < len(m) and 0 <= c < len(m[0])):
                continue
            if m[r][c] == '#': continue
            q.append((d+1, (r,c)))
    dmaps[st]=dmap
S_dmap = dmaps[(sr,sc)]
L = 26501365
olum = 0
um = 0

diag_memoize = {}
edge_memoize = {}

for r in range(R):
    for c in range(R):
        if (r,c) not in S_dmap: continue
        if S_dmap[(r,c)] % 2 == L % 2: um += 1
        for dir in  (0,0), (0,R-1), (R-1,0), (R-1,R-1):
            dum = 0
            d1 = S_dmap[dir] + 2
            d2 = dmaps[dir][(r,c)]
            if (d1+d2) in diag_memoize:
                um += diag_memoize[d1+d2]
                continue
            for i in range(0, L+1, R):
                d = d1+d2+i
                if d > L: break
                if d%2 == L%2: dum+= i//R + 1
            diag_memoize[d1+d2]=dum
            um += dum

        for dir in (sr,0), (sr,R-1), (0,sc), (R-1, sc):
            d1 = S_dmap[dir] + 1
            d2 = dmaps[dir][(r,c)]
            dum = 0
            if (d1+d2) in edge_memoize:
                um += edge_memoize[d1+d2]
                continue
            for i in range(0, L+1, R):
                d = d1+d2+i
                if d > L: break
                if d%2 == L%2: dum+=1
            edge_memoize[d1+d2]=dum
            um += dum

        print(r,c, um - olum)
        olum = um
ans(um)
```
