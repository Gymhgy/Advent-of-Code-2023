```python
sc = [0]
hare = 0
tort = 0
end = 1000000000

r = [''.join(x) for x in zip(*s.split())]

i = 0
stage = 1
clen = 1

while i < end:
    for _ in range(4):
        r = ['#'.join([('O'*c.count('O')).ljust(len(c), '.') for c in l.split('#')]) for l in r]
        r = [''.join(x) for x in zip(*r)][::-1]
    sc.append(hash(''.join(r)))
    i+=1
    if stage == 1:
        hare+=1
        if hare % 2 == 0: tort+=1
        if sc[tort] == sc[hare]:
            stage = 2
            tort = -1
    elif stage == 2:
        hare += 1
        tort += 1
        if i ==20: print(sc)
        if sc[tort] == sc[hare]:
            stage = 3
            hare = tort+1
    elif stage == 3:
        clen+=1
        hare += 1
        if sc[tort] == sc[hare]:
            #Yo we found our cycle
            i += clen * ((end - i)//clen)
            stage = 4


ans(sum((len(l)-j) for l in r for j,c in enumerate(l) if c=='O'))
```
