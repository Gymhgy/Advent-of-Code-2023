```python
from dataclasses import dataclass

@dataclass
class Light:
    row: int
    col: int
    dr: int
    dc: int

M = s.split()
def sim(start):
    S = []
    S.append(start)
    lit = set()
    been = set()
    while S:
        copy = list(S)
        for l in copy:
            if (l.row,l.col,l.dr,l.dc) in been \
            or not (0<=l.row<len(M) and 0<=l.col<len(M[0])):
                copy.remove(l)
                continue
            c = M[l.row][l.col]
            lit.add((l.row, l.col))
            been.add((l.row, l.col, l.dr, l.dc))
            if c == '/':
                l.dr, l.dc = -l.dc, -l.dr
            elif c == '\\':
                l.dr, l.dc = l.dc, l.dr
            elif c == '|' and l.dr == 0:
                l.dr, l.dc = 1, 0
                copy.append(Light(l.row-1, l.col, -1, 0))
            elif c == '-' and l.dc == 0:
                l.dr, l.dc = 0, 1
                copy.append(Light(l.row, l.col-1, 0, -1))
            l.row+=l.dr
            l.col+=l.dc

        S = copy
    return len(lit)
ans(sim(Light(0,0,0,1)))
```

```python
rs = len(M)
cs = len(M[0])
lr = max(max(sim(Light(i,0,0,1)), sim(Light(i,cs-1,0,-1))) for i in range(rs))
tb = max(max(sim(Light(0,i,1,0)), sim(Light(rs-1,i,-1,0))) for i in range(cs))
ans(max(lr, tb))
```
